```python
def load_batch_data(source_name:str, table_name:str):
    """
    Loads the data from the table_name to a bronze location

    Args:
        source_name (str): The name of the source
        table_name (str): The name of the table to load

    Returns:
        None
    """

    spark.table(f"samples.{source_name}.{table_name}").write.mode("overwrite").saveAsTable(f"workspace.bronze.{source_name}_{table_name}")
```
