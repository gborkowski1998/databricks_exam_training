```python
def load_batch_data(source_catalog_name:str, target_catalog_name:str, bronze_schema:str, source_name:str, table_name:str):
    """
    Loads the data from the table_name to a bronze location

    Args:
        source_catalog_name (str): The name of the source catalog
        target_catalog_name (str): The name of the target catalog
        bronze_schema (str): The name of the bronze schema
        source_name (str): The name of the source
        table_name (str): The name of the table

    Returns:
        None
    """

    spark.table(f"{source_catalog_name}.{source_name}.{table_name}").write.mode("overwrite").saveAsTable(f"{target_catalog_name}.{bronze_schema}.{source_name}_{table_name}")
```
