```python
# Generate new data
landing_path, checkpoint_path, schema_path = generate_streaming_data(source_name='wanderbricks', table_name='bookings', num_files=10, num_rows_per_file=1000)

bronze_df = (
    spark.readStream.format("cloudFiles") \
    .option("cloudFiles.format", "json") \
    .option("cloudFiles.schemaLocation", schema_path) \
    .option("cloudFiles.inferColumnTypes", "true") \
    .load(landing_path) \
    .writeStream \
    .format("delta") \
    .option("checkpointLocation", checkpoint_path) \
    .trigger(once=True) \
    .table("workspace.bronze.bookings")
)
```
