```python
bookings_df = spark.table("workspace.bronze.wanderbricks_bookings")
users_df = spark.table("workspace.bronze.wanderbricks_users")
properties_df = spark.table("workspace.bronze.wanderbricks_properties")

main_df = bookings_df.alias("bookings") \
                     .filter("status = 'confirmed'") \
                     .join(users_df.alias("users"), "user_id", "left") \
                     .join(properties_df.alias("properties"), "property_id", "left") \
                     .select("bookings.booking_id",
                             "bookings.check_in",
                             "bookings.check_out",
                             "bookings.status",
                             "bookings.total_amount",
                             "users.user_id",
                             "users.name",
                             "users.country",
                             "users.user_type",
                             "properties.property_id",
                             "properties.title") \
                     .withColumnRenamed("title", "Hotel_name")

main_df.write.mode("overwrite").saveAsTable("workspace.silver.wanderbricks_main")
```
